```python
df['Subscribers'] = df['Subscribers'].astype(float)
```

```python
df['Video Uploads'] = df['Video Uploads'].astype(float)
```

```python
df['Subscribers'] = df['Subscribers'].replace(np.nan,df['Subscribers'].mean())
```

```python
df['Video Uploads'] = df['Video Uploads'].replace(np.nan,df['Video Uploads'].mean())
```

```python
df['Rank'] = df['Rank'].str[0:-2]
```

```python
df['Rank'].str.replace(',','').astype(int)
```

Data Cleaning [ Grade Column ]

```python
df['Grade'].unique()
```

```python
df[df['Grade'] == '\xa0'].sum()
```

```python
df['Grade'] = df['Grade'].replace('\xa0','A',regex=True)
```

```python
df['Grade'] = df['Grade'].map({'A++ ':5, 'A+ ': 4, 'A ':3, 'A- ':2, 'B+ ':1}).astype(int)
```
